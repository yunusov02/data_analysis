```python
guassian_numbers = np.random.normal(size=1000)
guassian_numbers

plt.hist(guassian_numbers, bins=30, color="green", alpha=0.7)
plt.title("Histogram of Gaussian Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
X = np.linspace(0, 2 * np.pi, 100)

F1 = 3 * np.sin(X)
F2 = 2 * np.cos(X)
F3 = 0.3 * np.sin(5 * X)

fig, ax = plt.subplots()
ax.plot(X, F1, label="3*sin(X)", color="red")
ax.plot(X, F2, label="2*cos(X)", color="blue")
ax.plot(X, F3, label="0.3*sin(5*X)", color="green")
ax.set_title("Multiple Functions Plot")
ax.set_xlabel("X")
ax.set_ylabel("F(X)")
ax.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(X, F1, label="3*sin(X)", color="red")
ax1.set_title("3*sin(X)")
ax1.set_xlabel("X")
ax1.set_ylabel("F1(X)")
ax1.legend()

ax2.plot(X, F2, label="2*cos(X)", color="blue")
ax2.set_title("2*cos(X)")
ax2.set_xlabel("X")
ax2.set_ylabel("F2(X)")
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)


axes[0, 0].plot(x, y1, color="red")
axes[0, 0].set_title("Sine Function")

axes[0, 1].plot(x, y2, color="blue")
axes[0, 1].set_title("Cosine Function")

axes[1, 0].plot(x, y3, color="green")
axes[1, 0].set_title("Tangent Function")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(3, 3, figure=fig)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)
y5 = np.log(x + 1)  # Adding 1 to avoid log(0)

# Create subplots with custom spans
ax1 = fig.add_subplot(gs[0, :])  # Spans full width of first row
ax2 = fig.add_subplot(gs[1, :-1])  # Spans first two columns of second row
ax3 = fig.add_subplot(gs[1:, -1])  # Spans last column of second and third rows
ax4 = fig.add_subplot(gs[2, 0])  # Single cell
ax5 = fig.add_subplot(gs[2, 1])  # Single cell

# Add your plots here
ax1.plot(x, y1)
ax1.set_title('Full Width')

ax2.plot(x, y2)
ax2.set_title('Two Columns')

ax3.plot(x, y3)
ax3.set_title('Two Rows')

ax4.plot(x, y4)
ax4.set_title('Bottom Left')

ax5.plot(x, y1 + y2)
ax5.set_title('Bottom Middle')

plt.tight_layout()
plt.show()
```
