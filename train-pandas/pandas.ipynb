```python
top5field = df[df['field'].isin(['Law', 'Business', 'Science', 'Entertainment', 'Medicine', 'Social Work'])].copy()
top5field.sample(10)
```

```python
fig = plt.figure(figsize=(5, 3))

boxplot = sns.boxplot(
    x=top5field['attractive'],
    y=top5field['field'],
    hue=top5field['gender'],
    orient='h',
    palette= {
        "female": "lightpink",
        "male": "lightgreen"
    }
)

boxplot.set_title('Attractive')
boxplot.set_xlabel("Attractive")
boxplot.set_ylabel("Field")
boxplot.grid()
```

```python
fig = plt.figure(figsize=(5, 3))

boxplot = sns.boxplot(
    x=top5field['attractive_o'],
    y=top5field['field'],
    hue=top5field['gender'],
    orient='h',
    palette= {
        "female": "lightpink",
        "male": "lightgreen"
    }
)

boxplot.set_title('Attractive')
boxplot.set_xlabel("Attractive")
boxplot.set_ylabel("Field")
boxplot.grid()
```
