* `age` - age

* `Pclass` - Passenger class  (service classs)

* `fare` - price ticket  0 - 512
* `cabin` - number kayuti
* `embarked` - port posadki
    * `C` - Cherbourg
    * `Q` - Queenstown
    * `S` - Southampton

* `SibSp` - Siblings/Spouses count of siblings and spouses
* `parch` - Parents/Children count of parents or childrens

```python
# 1. Создаем сводную таблицу: Размер семьи vs Класс (или другой признак)
# Значения внутри — среднее по колонке Survived (т.е. вероятность выживания)
family_survival_pivot = df.pivot_table(index='FamilySize', 
                                       columns='Pclass', 
                                       values='Survived', 
                                       aggfunc='mean')

# 2. Рисуем Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(family_survival_pivot, annot=True, cmap='RdYlGn', fmt=".2f")

plt.title('Вероятность выживания в зависимости от размера семьи и класса')
plt.xlabel('Класс (Pclass)')
plt.ylabel('Размер семьи (FamilySize)')
plt.show()
```
