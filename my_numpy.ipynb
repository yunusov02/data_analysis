### Arange

**`arange([start] [stop] [step], [dtype=None])`**

```python
import numpy as np
```

```python
a = np.arange(1, 10)
print(a)
```

```python
x = range(1, 10)
print(x) # x is iterator, not an array
print(list(x)) # convert to list to see the values
```

```python
x = np.arange(0.9, 10.5, 2.3, int)
print(x)
```

### Linspace

**`linspace(start, stop num=50, endpoint=True, retstep=False)`**

```python
print(np.linspace(1, 10, 5)) # 5 value from 1 to 10, equally spaced
print(np.linspace(1, 10, 5, endpoint=False)) # exclude the endpoint (10)
print(np.linspace(1, 10, 5, retstep=True)) # return the step size as well
```

```python
dt = np.dtype([('density', np.int32)])

x = np.array([(345,), (456,), (654,) ], dtype=dt)

print(x)

print(repr(x))
```

```python
dt = np.dtype([
    ('country', np.str_, 20), # ('country', "U20"), 
    ('density', np.int32),
    ('area', np.int32),
    ('population', np.int32)
])

population_table = np.array([
    ('Netherlands', 393, 41526, 16928800),
    ('Belgium', 337, 30510, 11007020),
    ('United Kingdom', 256, 243610, 62262000),
    ('Germany', 233, 357021, 81799600),
    ('Liechtenstein', 205, 160, 32842),
    ('Italy', 192, 301230, 59715625),
    ('Switzerland', 177, 41290, 7301994),
    ('Luxembourg', 173, 2586, 512000),
    ('France', 111, 547030, 63601002),
    ('Austria', 97, 83858, 8169929),
    ('Greece', 81, 131940, 11606813),
    ('Ireland', 65, 70280, 4581269),
    ('Sweden', 20, 449964, 9515744),
    ('Finland', 16, 338424, 5410233),
    ('Norway', 13, 385252, 5033675)
], dtype=dt)

print(population_table['country'][2:5])
print(population_table['density'])
```

```python
# To save csv file

np.savetxt(
    "population_table.csv",
    population_table,
    fmt="%s;%d;%d;%d",
    delimiter=";"
)
```

```python
x = np.genfromtxt(
    "population_table.csv",
    dtype=dt,
    delimiter=";"
)
print(type(x))

# print(x)

print(x['country'])
print(x['population'])
```

```python
dt_products = np.dtype([
    ('id', np.int32),
    ('price', np.int32),
    ('name', np.str_, 100)
])

product_data = [
    (1, 25, "Wireless Mouse"),
    (2, 89, "Mechanical Keyboard"),
    (3, 45, "USB-C Hub"),
    (4, 299, "Ergonomic Chair")
]

products = np.array(product_data, dtype=dt_products)

new_item = np.array([(5, 120, "Portable SSD")], dtype=dt_products)
products = np.append(products, new_item)

print(products)
print(f"Product Names: {products['name']}")

```
